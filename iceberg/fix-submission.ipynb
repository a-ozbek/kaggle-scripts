```python
df_sub = pd.read_csv('./submissions/sub5-6-7-8-10-11-12-18-19.csv')
df_sub
```

```python
fig = plt.figure(figsize=(10, 6))
df_sub['is_iceberg'].hist().plot()
```

```python
cond1 = (df_test['band_1'].apply(lambda x: np.max(x)) > 20)
cond2 = (df_test['band_2'].apply(lambda x: np.max(x)) > 5)
cond3 = (df_test['band_2'].apply(lambda x: np.std(x)) > 7)
cond4 = (df_test['band_2'].apply(lambda x: np.std(x)) > 5)

df_sub_ = df_sub[cond1 & cond2 & cond3 & cond4]
print len(df_sub_)
df_sub_['is_iceberg'].hist().plot()
```
